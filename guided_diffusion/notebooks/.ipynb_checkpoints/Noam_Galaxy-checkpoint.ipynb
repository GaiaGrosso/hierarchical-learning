```python
import os
import numpy as np
import torch as th
from guided_diffusion import dist_util, logger
from guided_diffusion.script_util import (
    NUM_CLASSES,
    model_and_diffusion_defaults,
    create_model_and_diffusion,
)
from PIL import Image
import datetime
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages  # Import PdfPages for creating a PDF

# Settings (replace with your own model path, data directory, and output path)
# model_path = "/home/nlevi/Noam/GalaxyDiff/guided-diffusion/256x256_diffusion_uncond.pt"


model_path = "/home/nlevi/Noam/GalaxyDiff/guided-diffusion/fine_tuned_models/model001000.pt"


use_ddim = False
output_dir = os.path.join(os.getcwd(), 'results', 'forw_back', datetime.datetime.now().strftime("%Y-%m-%d-%H-%M-%S-%f"))
num_samples = 1
batch_size = 1
class_cond = False
time_series = [25, 50, 100, 150, 200, 250 ,500,700, 999]
clip_denoised = True
image_size = 256  # Set to your model's expected input size
device = 'cuda' if th.cuda.is_available() else 'cpu'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Load model and diffusion
model, diffusion = create_model_and_diffusion(**model_and_diffusion_defaults())
model.load_state_dict(dist_util.load_state_dict(model_path, map_location="cpu"))
model.to(device)
model.eval()

# Optionally convert to FP16 if needed
if use_ddim:
    model.convert_to_fp16()

# Load a specific image as input
def load_specific_image(image_path, image_size):
    img = Image.open(image_path).convert("RGB").resize((image_size, image_size))  # Ensure RGB
    img = np.array(img).astype(np.float32) / 127.5 - 1.0  # Normalize to [-1, 1]
    img = th.from_numpy(img).permute(2, 0, 1).unsqueeze(0)  # Add batch and channel dimension
    return img.to(device)

# Sampling loop without distributed functions
def sample_loop(model, diffusion, start_image, step_reverse):
    all_images = []

    while len(all_images) * batch_size < num_samples:
        batch_start = start_image.repeat(batch_size, 1, 1, 1)  # Repeat for batch size
        t_reverse = diffusion._scale_timesteps(th.tensor([step_reverse])).to(device)
        batch_noisy = diffusion.q_sample(batch_start, t_reverse)

        model_kwargs = {}
        sample_fn = diffusion.p_sample_loop_forw_back if not use_ddim else diffusion.ddim_sample_loop_forw_back
        sample = sample_fn(
            model,
            (batch_size, 3, image_size, image_size),
            step_reverse=step_reverse,
            noise=batch_noisy,
            clip_denoised=clip_denoised,
            model_kwargs=model_kwargs,
        )
        
        # Process and save the generated sample
        sample = ((sample + 1) * 127.5).clamp(0, 255).to(th.uint8)
        sample = sample.permute(0, 2, 3, 1).cpu().numpy()

        all_images.append((step_reverse, sample))  # Save step_reverse with image

        print(f"Created {len(all_images) * batch_size} samples")

    return all_images

# Create a PDF with all images
def save_images_to_pdf(images, output_pdf_path):
    with PdfPages(output_pdf_path) as pdf:
        for step_reverse, img in images:
            fig, ax = plt.subplots(figsize=(6, 6))
            ax.imshow(img[0])
            ax.set_title(f"Step Reverse: {step_reverse}")
            ax.axis('off')
            pdf.savefig(fig)  # Save the current figure to the PDF
            plt.close(fig)  # Close the figure to free memory

# Example usage in the notebook
# image_path = "/home/nlevi/Noam/GalaxyDiff/guided-diffusion/datasets/e1b-chandrawebb2_ngc346_comp-web.width-1280.jpg"

image_path = "/home/nlevi/Noam/GalaxyDiff/guided-diffusion/datasets/e4b-chandrawebb2_m74_comp-web.width-1280.jpg"
start_image = load_specific_image(image_path, image_size)




# Collect images for each step_reverse
all_generated_images = []
for step_reverse in time_series:
    print(f"Sampling for step_reverse = {step_reverse}")
    all_generated_images.extend(sample_loop(model, diffusion, start_image, step_reverse))

# Save all images to a single PDF
pdf_path = os.path.join(output_dir, "generated_samples.pdf")
save_images_to_pdf(all_generated_images, pdf_path)
print(f"All images saved to {pdf_path}")
```
